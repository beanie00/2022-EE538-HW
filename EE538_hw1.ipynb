# Problem2

```python
import numpy as np
import matplotlib.pyplot as plt


def prob2(vi, theta, tau, n):
    spike_time = 0
    x = np.linspace(0, 100, n)
    y = np.linspace(0, 100, n)

    for i in range(1, n) :
      if y[i-1] > theta and y[i-1] < 55: 
        y[i] = 55
        spike_time = i
      else :
        y[i] = vi * (1 - np.exp((-x[(i - spike_time)]/tau)))

    plt.figure()
    plt.title("[prob2] vi: "+str(vi)+" theta: "+str(theta)+" tau: "+str(tau))
    plt.plot(x, y)
    plt.xlabel('$t$')
    plt.ylabel('$u(t)$')

    plt.show()

if __name__ == '__main__':
    prob2(12, 10, 10, 1000)
    prob2(12, 10, 20, 1000)
    prob2(100, 20, 15, 1000)
    prob2(50, 20, 20, 1000)
    prob2(100, 15, 50, 1000)
```

# Problem 5

```python
def prob5(theta, tau, n):
    n = 1000
    x = np.linspace(0, 100, n)
    y = -1/(tau*np.log(1-theta/x))

    plt.figure()
    plt.title("[prob5] theta: "+str(theta)+" tau: "+str(tau))
    plt.plot(x, y)
    plt.xlabel('$Vi$')
    plt.ylabel('$number of spikes [t]$')

    plt.show()

if __name__ == '__main__':
    prob5(10, 20, 1000)
    prob5(20, 15, 1000)
    prob5(15, 50, 1000)
```

# Problem6

```python
np.warnings.filterwarnings('ignore')

def prob6(mu, tau, n):
    sigma1 = mu/10
    sigma2 = mu/20
    x = np.linspace(1, 100, n)
    theta1 = np.random.normal(mu, sigma1, size=300)
    theta2 = np.random.normal(mu, sigma2, size=300)

    plt.figure()
    plt.title("[prob6] mu: "+str(mu)+" tau: "+str(tau)+ " sigma = mu/10")
    for i, t in enumerate(theta1):
      y = -1/(tau*np.log(1-t/x))
      plt.plot(x, y)
      plt.xlabel('$Vi$')
      plt.ylabel('$number of spikes [t]$')
    plt.show()

    plt.figure()
    plt.title("[prob6] mu: "+str(mu)+" tau: "+str(tau)+ " sigma = mu/20")
    for i, t in enumerate(theta2):
      y = -1/(tau*np.log(1-t/x))
      plt.plot(x, y)
      plt.xlabel('$Vi$')
      plt.ylabel('$number of spikes [t]$')
    plt.show()

if __name__ == '__main__':
    prob6(20, 10, 1000)
    prob6(50, 2, 1000)
```

# Problem7

```python
np.warnings.filterwarnings('ignore')

def prob7(mu, theta, n):
    sigma1 = mu/10
    sigma2 = mu/20
    x = np.linspace(1, 100, n)
    tau1 = np.random.normal(mu, sigma1, size=300)
    tau2 = np.random.normal(mu, sigma2, size=300)

    plt.figure()
    plt.title("[prob7] mu: "+str(mu)+" theta: "+str(theta)+" sigma = mu/10")
    for i, t in enumerate(tau1):
      y = -1/(t*np.log(1-theta/x))
      plt.plot(x, y)
      plt.xlabel('$Vi$')
      plt.ylabel('$number of spikes [t]$')
    plt.show()

    plt.figure()
    plt.title("[prob7] mu: "+str(mu)+" theta: "+str(theta)+" sigma = mu/20")
    for i, t in enumerate(tau2):
      y = -1/(t*np.log(1-theta/x))
      plt.plot(x, y)
      plt.xlabel('$Vi$')
      plt.ylabel('$number of spikes [t]$')
    plt.show()

if __name__ == '__main__':
    prob7(20, 20, 1000)
    prob7(50, 10, 1000)
```
